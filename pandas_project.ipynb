```python
#concatenate and merge dataframes (joins)
df1 = pd.DataFrame({"ID" : [1,2,3] , "Name" : ['a' ,'b', 'c']})
df1

```

```python
df2 = pd.DataFrame({"ID" :[1,2,4,3] , "score" : [ 33,44,5,5]})
df2
```

```python
pd.concat([df1 , df2] , axis = 0 ) #vertical / row level / top
```

```python
pd.concat([df1 , df2] , axis = 1) 3 horizontal level
```

```python
# merge = join
pd.merge(df1 , df2 , how = "right" , on = "ID")
```

```python
pd.merge(df1 , df2 , how = "inner" , left_on = "ID" , right_on = "ID")
```
